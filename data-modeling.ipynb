```python
import pandas as pd 
```

```python
df = pd.read_csv("OnlineRetail.csv", sep=',', encoding='latin1')
```

```python
df.head()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.head()
```

```python
datetime_dim = df[['InvoiceDate']].reset_index(drop=True)
datetime_dim['InvoiceDate'] = datetime_dim['InvoiceDate']
datetime_dim['invoice_day'] = datetime_dim['InvoiceDate'].dt.day
datetime_dim['invoice_month'] = datetime_dim['InvoiceDate'].dt.month
datetime_dim['invoice_year'] = datetime_dim['InvoiceDate'].dt.year
```

```python
datetime_dim['datetime_id'] = datetime_dim.index
```

```python
datetime_dim.head()
```

```python
# Check if 'StockCode' is present in df
if 'StockCode' in df.columns:
    # Create product_dim with 'StockCode' and reset the index
    product_dim = df[['StockCode']].reset_index(drop=True)

    # Check if 'Description' and 'UnitPrice' are present in product_dim
    if all(col in product_dim.columns for col in ['Description', 'UnitPrice']):
        # Update product_dim to contain only 'Description' and 'UnitPrice'
        product_dim = product_dim[['Description', 'UnitPrice']]
    else:
        print("Error: 'Description' and 'UnitPrice' columns are not present in product_dim.")
else:
    print("Error: 'StockCode' column is not present in df.")
```

```python
product_dim = df[['StockCode']].drop_duplicates().reset_index(drop=True)
```

```python
product_dim = product_dim.merge(df[['StockCode', 'Description', 'UnitPrice']], on='StockCode', how='left')
```

```python
product_dim.head()
```

```python
# Create customer_dim with unique 'CustomerID'
customer_dim = df[['CustomerID']].drop_duplicates()

# Add 'Country' to customer_dim if it exists, else fill NaN values with 'Unknown'
customer_dim['Country'] = df.groupby('CustomerID')['Country'].first()

# Reset the index and print the resulting DataFrame
customer_dim.reset_index(drop=True, inplace=True)
print(customer_dim)
```
